```python
# Figure and axes for the univariate analysis
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(10, 15)) #Making subplots

# List of variables to plot
target_variables = [
    "Data_Operations",
    "Data_Bytes",
    "Disk_Reading",
    "Received_Bytes",
    "Bytes_Sent",
    "Network_packets_sent",
    "Network_packets_received",
    "Time_processor"
]

# Generating histograms
for ax, var in zip(axes[:, 0], target_variables):
    sns.histplot(data_source[var], ax=ax, bins=30, kde=True)
    ax.set_title(f"Histogram of {var}")
    ax.set_ylabel("Frequency")
    
# Generating box plots
for ax, var in zip(axes[:, 1], target_variables):
    sns.boxplot(y=data_source[var], ax=ax)
    ax.set_title(f"Box Plot of {var}")
    
plt.tight_layout()
plt.show()
```
